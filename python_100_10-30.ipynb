두 개를 합쳐 학생의 이름이 key이고 value가 수학 점수인 딕셔너리를 출력해주세요.  

입력  
Yujin Hyewon  
70 100  

출력
```python
{'Yujin': 70, 'Hyewon': 100}
```
#이름을 input과 split함수를 이용해 공백으로 구분해 입력받는다.  
점수는 공백으로 구분해 정수형으로 입력받는다.  
결과는 dict()함수를 이용해 key와 value를 함께 출력한다.

## 28. 2-gram  
2-gram이란 문자열에서 2개의 연속된 요소를 출력하는 방법입니다.  
예를 들어 'Python'을 2-gram으로 반복해 본다면 다음과 같은 결과가 나옵니다.   

Py  
yt  
th  
ho  
on  

입력으로 문자열이 주어지면 2-gram으로 출력하는 프로그램을 작성해 주세요.

````python
a = input()

for i in range(len(a) - 1):
    print(a[i], a[i+1], sep='')
````
