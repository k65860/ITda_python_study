### 10. 세 수 A, B, C가 주어졌을 때, 위의 네 가지 값을 구하는 프로그램을 작성하시오.  첫째 줄에 (A+B)%C, 둘째 줄에 ((A%C) + (B%C))%C, 셋째 줄에 (A×B)%C, 넷째 줄에 ((A%C) × (B%C))%C를 출력한다.

```python
A, B, C = map(int,input().split())
print((A + B) % C)
print((A % C) + (B % C) % C)
print((A * B) % C)
print((A % C) * (B % C) % C)
```

### 2. 시험 점수를 입력받아 90 ~ 100점은 A, 80 ~ 89점은 B, 70 ~ 79점은 C, 60 ~ 69점은 D, 나머지 점수는 F를 출력하는 프로그램을 작성하시오.  시험 점수는 0보다 크거나 같고, 100보다 작거나 같은 정수이다.

```python
a = int(input())

if a >= 90 and a <= 100:
    print('A')
elif a >= 80 and a <= 89:
    print('B')
elif a >= 70 and a <= 79:
    print('C')
elif a >= 60 and a <= 69:
    print('D')
else:
    print('F')
```

### 3. 연도가 주어졌을 때, 윤년이면 1, 아니면 0을 출력하는 프로그램을 작성하시오.

윤년은 연도가 4의 배수이면서, 100의 배수가 아닐 때 또는 400의 배수일 때이다.

예를 들어, 2012년은 4의 배수이면서 100의 배수가 아니라서 윤년이다. 1900년은 100의 배수이고 400의 배수는 아니기 때문에 윤년이 아니다. 하지만, 2000년은 400의 배수이기 때문에 윤년이다.

```python
a = int(input())

if a % 4 == 0 and a % 100 != 0:
    print(1)
elif a % 400 == 0:
    print(1)
else:
    print(0)
```

### 5. 알람 시계
